```python
import requests
import pandas as pd

# Replace with your actual Bearer Token from your Twitter Developer App
BEARER_TOKEN = 'AAAAAAAAAAAAAAAAAAAAAFTIvQEAAAAACqwbJvmFj6Wv340brTUOsgbrG3k%3D09C8y1453vBNiD9SxsfcnbWef4NpiN7yRUtFMlEyHynLywgHdB'

# The search query you want to use
search_query = "Nigga"

# Number of tweets to retrieve (max is 100 for recent search)
no_of_tweets = 50

# Endpoint URL for recent search (Twitter API v2)
search_url = "https://api.twitter.com/2/tweets/search/recent"

# Parameters for the API request
query_params = {
    'query': search_query,
    'tweet.fields': 'author_id,created_at,public_metrics,source,text',
    'max_results': no_of_tweets,
    'start_time': '2024-03-01T00:00:00Z',  # Start date (inclusive)
    'end_time': '2024-05-31T23:59:59Z',    # End date (inclusive)
    'lang': 'en'  # Filtering tweets in English
}

# Headers with Bearer Token for authorization
headers = {"Authorization": f"Bearer {BEARER_TOKEN}"}

# Making the request to the Twitter API
response = requests.get(search_url, headers=headers, params=query_params)

# Checking if the request was successful
if response.status_code == 200:
    tweets = response.json().get('data', [])
    if tweets:
        # Extracting desired attributes from each tweet
        attributes_container = [
            [
                tweet['author_id'],
                tweet['created_at'],
                tweet['public_metrics']['like_count'],
                tweet['source'],
                tweet['text']
            ]
            for tweet in tweets
        ]

        # Creating column names for the DataFrame
        columns = ["User ID", "Date Created", "Number of Likes", "Source of Tweet", "Tweet"]

        # Creating the DataFrame
        tweets_df = pd.DataFrame(attributes_container, columns=columns)
        print(tweets_df.head())  # Display the first few rows of the DataFrame
    else:
        print("No tweets found for the given query and date range.")
else:
    print(f"Status Failed On, {response.status_code} {response.text}")
```
